#### Passos para criar/ treinar um modelo
- Passo 1: Definir se é Classificação ou Regressão
- Passo 2: Escolher as Métricas para avaliar o Modelo
- Passo 3: Escolher quais Modelos vamos usar/testar
- Passo 4: Treinar os Modelos e Testar
- Passo 5: Comparar os Resultados dos modelos e escolher o vencedor
- Passo 6: Analisar o melhor modelo mais a fundo
- Passo 7: Fazer ajustes no melhor modelo 

#### (nota pessoal)
#### Explicação passo 1
- Regressão:
Quando o resultado buscado é um preço
- Classificação:
Quando o resultado é definido em categorias (EX.: Modelo de machinelearning para descobrirdiagnósticos de doenças, a pessoa vai ter a doença ou não vai ter a doença)

#### Explicação passo 2
###### MODELO R^2:
- É dados de 0 a 1 -> quanto mais próximo de 1 melhor
- Explicação simples: Mede "o quanto" dos valores o modelo consegue explciar
- EX.: 92% significa que o modelo consegue explicar 92% da variância dos dados a partir das informações que você dá para ele

###### MODELO RSME (erro quadrático Médio ou Raiz do Eroo Quadrático Médio)
- Pode ser qualquer valor
- Mede o quanto o modelo erra


#### Explicação passo 3
##### Modelos de regressão
- Regressão linear: Traça uma reta que pega a média da distância entre os pontos
- Random Forest Regressor: Ele funciona criando várias árvores de decisão e combinando as previsões delas. Cada árvore faz uma estimativa, e o modelo calcula a média dessas previsões para obter um resultado mais preciso e estável. (tipo o akinator)
- Extra trees: Faz a mesma coisa que o random forest, mas, enquanto o radom forest escolhe a melhor pergunta, o extra trees escolhe uma pergunta aleatória

#### Explicação passo 4
##### Separação da Base de dados em Treino e Teste
- Separar os dados aleatoriamente em 2 conjuntos: Treino e teste (A maior parte vai ser para treino)
- O treino são os dados que o odelo vai usar para aprender
- O teste são os dados que serão utilizados para analisar se o modelo aprendeu bem
- EX.: 80% dados treino e 20% dados teste.
- Avaliar sempre o resultado nos testes, para não correr o risco de overfitting  (O modelo começa a prever errado por ter ficado específico de mais, não conseguindo se adaptar a dados novos)

#### Explicação passo 5
- Calcular o R^2 e o RSME para cada modelo
- Escolher 1 métrica para ser a principal, digamos R^2
- Nesse caso, o modelo com maior R^2 vai ser considerado o melhor modelo
- Usar o RSME para critério de desempate ou para comparar modelos com R^2 parecidos
##### MAS, esses não são os únicos critérios:
- Um modelo que demore 1hr para treinar e fazer previsões é pior do que um modelo que demore 5min
- Um modelo que precise de menos informações para funcionar (mais simples) é melhor do que um modelo mais complexo

#### Explicação passo 6
- Depois do modelo pronto e escolhido o melhor modelo, analisar como o melhor modelo funciona
- Identificar a importância de cada feature para ver oportunidades de melhoria

- Ex: Se uma feature/coluna não é utilizada pelo modelo, pode testar retirar essa feature/coluna do modelo e ver o resultado
- O resultado pode melhorar ou piorar, mas lembrar de sempre olhar:
- Métricas escolhidas (R^2 e RSME)
- Velocidade do modelo
- Simplicidade do modelo

#### Explicação passo 7
##### Testar a cada mudança que é feita no modelo para fazer um ajuste final

- Ver se as features identificadas no último passo pode ser retirada
- A cada etapa, treinar e testar o modelo, sempre comparando com o resultado original e o resultado anterior
##### Objetivo
- Encontrar uma possível melhoria ao modelo (inclusive de previsão -> maior R^2 ou menor RSME)
- Ver se foi possível chegar no mesmo resultado ou em um resultado muito próximo com um modelo mais simples e mais rápido
- Fazer outros testes que sejam desejados ou esteja planejado na
